```python
# Specify the path to your web driver executable
driver_path = 'C:\Program Files (x86)\chromedriver.exe'  # Replace with the correct path to chromedriver.exe

keywordlist = ['pest','safety','policy']

# Specify keyword
keyword = 'pest'
# Replace 'url' with the website URL you want to scrape
base_url = "https://www.myskillsfuture.gov.sg/content/portal/en/training-exchange/course-directory.html?fq=Course_Supp_Period_To_1%3A%5B2023-08-02T00%3A00%3A00Z%20TO%20*%5D&fq=IsValid%3Atrue&q="
bbaseurl = base_url + keyword + "&start="

# Start value for the "&start=" parameter
start_value = 0

# Create an empty list to store the course titles
course_title_list = []
course_title_list_url = []

# Create an empty list to store the course content
course_content = []
while True:
 
    url = bbaseurl + str(start_value)

    # Create a web driver instance (e.g., Chrome)
    driver = webdriver.Chrome(driver_path)
    driver.get(url)

    # Introduce a load time between requests (e.g., 3 seconds)
    time.sleep(3)

    specific_elements_groups = driver.find_elements(By.CLASS_NAME, 'page-content-body-holder')
    for specific_elements_group in specific_elements_groups:
       specific_elements = specific_elements_group.find_elements(By.CLASS_NAME, 'col')

    if not specific_elements:
       # If the list is empty (element not found), break the loop
       print("Specific element not found. Exiting loop.")
       break

    # Find sections that contain the courses
    sections = driver.find_elements(By.CLASS_NAME, 'page-content-body-holder')

    # Find anchor elements (links) within each section
    for section in sections:
        col_elements = section.find_elements(By.CLASS_NAME, 'card-title')

        # Extract and add the text content of each course title to the list
        for element in col_elements:
            course_title_list.append(element.text.strip())

    # Find all sections that contain the links

    groups = driver.find_elements(By.CLASS_NAME, 'page-content-body-holder')

    for group in groups:
        sections = group.find_elements(By.CLASS_NAME, 'card-title')

    # Find anchor elements (links) within each section
    for section in sections:
       link_elements = section.find_elements(By.TAG_NAME, 'a')

       # Extract and print the href links for each section
       for link in link_elements:
           href = link.get_attribute('href')
           if href and href.startswith('http'):  # Filter out relative links
                course_title_list_url.append(href)

    # Increment the start_value by 24 for the next page
    start_value += 24
    
    driver.close()

def extract_text_from_element(driver, url, element_selector):
    driver.get(url)
    time.sleep(7)
    element = driver.find_element(By.CSS_SELECTOR, element_selector)
    return element.text if element else None

# CSS selector of the element containing the paragraph you want to extract
element_selector = 'p[data-bind="html : courseDetail.courseContent"]'

for url in course_title_list_url:
    coursecontent = extract_text_from_element(driver, url,element_selector)
    course_content.append(coursecontent)

driver.close()

def close_all_chrome_windows():
    # Initialize the ChromeDriver
    driver = webdriver.Chrome(executable_path='path_to_chromedriver')  # Replace 'path_to_chromedriver' with the actual path to chromedriver.exe

    # Get the list of all open windows/tabs
    window_handles = driver.window_handles

    # Close all windows
    for handle in window_handles:
        driver.switch_to.window(handle)
        driver.close()

    # Close the driver
    driver.quit()

if __name__ == "__main__":
    close_all_chrome_windows()    

course_title_list_url
courselist = list(zip(course_title_list,course_title_list_url,course_content))
df = pd.DataFrame(courselist, columns=['Course Title', 'Course Url','Course Content'])
excel_filename = 'output.xlsx'
df.to_excel(excel_filename, index=False)
```
