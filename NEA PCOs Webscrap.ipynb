```python
# Scrap for UEN based on CompanyName

import time
import random
import requests
from bs4 import BeautifulSoup

# Add a new column 'UEN' to the DataFrame
df['UEN'] = ''

for index, row in df.iterrows():
    company_name = row['CompanyName']
    search_query = "https://www.google.com.sg/search?q=" + company_name + " uen opengovsg.com"

    header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "en-US,en;q=0.9,lt;q=0.8,et;q=0.7,de;q=0.6",
}
    response = requests.get(search_query, headers = header)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Convert parsed HTML to string
    html_string = str(soup)

    # Find the position of the target term
    target_term = "https://opengovsg.com/corporate/"
    start_index = html_string.find(target_term)

    if start_index != -1:
        # Extract the next 10 characters after the target term
        extracted_text = html_string[start_index + len(target_term):start_index + len(target_term) + 10]
        df.at[index, 'UEN'] = extracted_text
    else:
        df.at[index, 'UEN'] = "UEN not found"
        
    # Add a random time delay between 5 and 20 seconds
    random_delay = random.uniform(5, 20)
    time.sleep(random_delay)
```
